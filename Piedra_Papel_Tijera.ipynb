"Piedra, Papel & Tijera", es un juego en el que dos participantes (en este caso, usuario ğŸ˜ y ordenador ğŸ’») tienen que elegir una de las tres opciones posibles entre "Piedra", "Papel" o "Tijera" con el fin de determinar un ganador segÃºn tres posibles combinaciones.

Las combinaciones posibles son:
Tijera gana a Papel, Papel gana a Piedra y Piedra gana a Tijera. Si el usuario y el ordenador eligen la misma opciÃ³n, se verifiquerÃ¡ un empate.

El jugador que gana la ronda obtiene un punto. El juego continua hasta que, en las varias rondas, uno de los dos jugadores acumule tres puntos y serÃ¡ entonces el vencedor del juego.ğŸ†ğŸ‘

Resumiendo:

- Tenemos 3 elementos: Piedra, Papel y Tijera;
- Cada partecipante, uno contra uno, elige uno de esos 3 elementos con el objetivo de ganar el adversario; 
- Combinaciones posibles: Tijera gana a Papel
                          Papel gana a Piedra
                          Piedra gana a Tijera

```python
import random # El modulo random genera al azar una de las tres opciones para el jugador "ordenador"


ordenador_puntos = 0 # Creamos las variables "ordenador_puntos" y "usuario_puntos" para que sea posible hacer un seguimiento de la puntuaciÃ³n.
usuario_puntos = 0
ronda = 0 # Esta variable nos permite hacer un seguimiento de las rondas.


while True: # Utilizamos un bucle 'while' para que el juego pueda continuar hasta que uno de los dos participantes alcance los 3 puntos.
    
    print("---------------------------------------------")
    print(f"------------------RONDA {ronda}--------------------")
    usuario = input("Bienvenid@ ğŸš€. Elige: Piedra, Papel o Tijera").lower() # Definimos la variable "usuario".Con el input, solicitamos al usuario que ingrese la opciÃ³n deseada entre piedra, papel o tijera. Aplicamos el metodo .lower() para que los datos ingresados con mayuculas no afecten el juego.

    opciones = ("piedra", "papel", "tijera") # Definimos las opciones posibles

    ordenador = random.choice(opciones) # Definimos la variable "ordenador". Se generarÃ¡ de forma aleatoria la opciÃ³n del ordenador.


    if usuario in opciones:  #Evaluamos (con el uso de if, elif, else) las 3 condiciones del usuario y del ordenador segÃºn las respuestas de los dos jugadores para que sea posible obtener un ganador. Evaluamos tambiÃ©n la condiciÃ³n de "Empate"
        
        
        if usuario == ordenador : #Es empate. Nadie gana puntos, no sumamos nada ni a ordenador_puntos ni a usuario_puntos
            print(f"Has elegido {usuario}. Es un empate, el ordenador ha elegido {ordenador}")
            
            
        elif (usuario == "piedra" and ordenador == "tijera") or (usuario == "papel" and ordenador == "piedra") or (usuario == "tijera" and ordenador == "papel") :
            #Gana usuario. Sumo 1 punto a usuario_puntos
            print(f"Has elegido {usuario}, ganaste!ğŸ’ª. El ordenador ha elegido {ordenador}")
            usuario_puntos += 1 #permite el conteo de uno a uno 
            print (f"Tu puntuaciÃ³n actual es: {usuario_puntos}")
            
        
        else:
            #sumo 1 punto a ordenador_puntos
            print(f"PerdisteğŸ˜¢. Has elegido {usuario} y el ordenador ha elegido {ordenador}")
            ordenador_puntos += 1
            print (f"La puntuaciÃ³n actual del ordenador es: {ordenador_puntos}")
            
        #Alguno de los jugadores ha alcanzado los 3 puntos?
        if usuario_puntos == 3 or ordenador_puntos == 3:
            break #Termina el juego
     
        
        print("---------------------------------------------")
        print(f"---------------FINAL RONDA {ronda}-----------------")
        ronda +=1 #Conteo de las rondas
    else:
        print(f"Ooops!ğŸ˜µ Has ingresado un valor invÃ¡lido. Elige entre: {opciones}")
        

if usuario_puntos == 3:
    print("*********************************************")
    print("ğŸ† Enhorabuena, has ganado el juego!ğŸ†")
    print("*********************************************")

elif ordenador_puntos == 3:
    print("OoooohğŸ˜­, has perdido, intentalo de nuevo!")

else:
    print("Ooops, estamos trabajando en eso, contacta con el administradorğŸ‘€")
```

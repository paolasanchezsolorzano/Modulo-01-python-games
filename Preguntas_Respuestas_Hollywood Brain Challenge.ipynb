Creamos dos listas de emojis y mejorar la experiencia del usuario.

```python
emojis_ganadores = [" la suerte está contigo 🫵", " muy bien 👍"," eres un crack 🤠", " WoW 😺", " Vamos, vamos!🕺", " Ole Ole 💃", " ¡como sabes!🙉", " clap, clap! 👏" ]
emojis_burlones = ["👻", "🤡","🙈", "🐣", "🙅‍♀️", "🥱", "🫢", "😪" ]
```

```python
def mostrar_pregunta(pregunta):
    print(pregunta["pregunta"])                              #Creamos función para mostrar las preguntas al usuario.
    for opcion in pregunta["opciones"]:
        print(opcion)
    respuesta_jugador = input("Adivina la opción correcta 🧐 (A, B, C, D) o escribe 'salir' ").upper() #input_usuario con .upper() para filtrar opciones.
    return respuesta_jugador

def seleccionar_pregunta(lista_preguntas):     #Obtenemos preguntas aleatorias con el random.choice.
    return random.choice(lista_preguntas)

def inicia_juego():                           #Función para iniciar juego.
    preguntas = base_datos
    intentos = 0
    aciertos = 0
    aciertos_max = 5      # Variable que marca el numero de intentos maximos.
    errores_max = 3       # Variable que marca el numero de fallos máximos.

    while aciertos < aciertos_max and intentos < errores_max: #Abrimos bucle.
        
        pregunta_actual = seleccionar_pregunta(preguntas)      # Definimos variable con la funcion que extrae las  preguntas y respuestas de la Base de Datos.
        respuesta_jugador = mostrar_pregunta(pregunta_actual)  # Definimos variable a del jugador.para capturar respuesta.
        
        
        if respuesta_jugador == "SALIR":
            print("Has salido del juego 👋")
            break

        if respuesta_jugador == pregunta_actual["respuesta_correcta"]:  # Usamos condicional para comprobar respuestas y sumar al marcador.
            aciertos += 1
            print(f'¡Bien!{random.choice(emojis_ganadores)} Llevas, {aciertos} aciertos.') # Añadimos lista de emojis con random.choice() para respuestas correctas.
        else:
            intentos += 1                                       
            print(f"Incorrecto {random.choice(emojis_burlones)}   La respuesta correcta era:", pregunta_actual["respuesta_correcta"])
            print("Llevas", intentos, "errores de 3 permitidos.")

    if aciertos == aciertos_max:                # Condicional anuncia ganador
        print("¡Perfecto! Has acertado 5 preguntas. Eres un Cinéfilo-Binge Watcher! 🥳😉")
        
    elif intentos == errores_max:               #Condicional anuncia perdedor
        print("Oh no, has perdido. 😞 Pero puedes volver a jugar. Tenemos muchas más preguntas para probar tus conocimientos! 😎")


inicia_juego() # Función que ejecuta el juego.
```
